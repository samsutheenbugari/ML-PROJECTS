#### Creating common function

```python
def runRandomforest(xtr,xts,ytr,yts):
    rfmodel=RandomForestClassifier(max_depth=7)
    rfmodel.fit(xtr,ytr)
    ytrp=rfmodel.predict(xts)
    print("acc:", accuracy_score(yts,ytrp))
    print("cmx:",confusion_matrix(yts,ytrp))
```

#### Checking

```python
%%time
runRandomforest(xtr,xts,ytr,yts)
```

```python
%%time
runRandomforest(xtr_transformed,xts_transformed,ytr,yts)
```

#### Creating loop for check optimum Columns

```python
for i in range(1,len(x.columns)):
    rfemodel=RFE(estimator=RandomForestClassifier(n_estimators=50,criterion='gini',max_depth=10),n_features_to_select=i)
    rfemodel.fit(xtr,ytr)
    xtr_transformedrfe=rfemodel.transform(xtr)
    xts_transformedrfe=rfemodel.transform(xts)

    print(xtr_transformedrfe.shape)
    print("num of features",i)
    runRandomforest(xtr_transformedrfe,xts_transformedrfe,ytr,yts)
    print("#########")
    
```
